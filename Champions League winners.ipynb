```python
import pandas as pd
import matplotlib.pyplot as plt
```

Unveiling Champions League Finalist History with Python
Football fans, this one's for you!

```python
table=pd.read_html("https://en.wikipedia.org/wiki/List_of_European_Cup_and_UEFA_Champions_League_finals")
print(len(table))
```

```python
print(table[2])
```

```python
#selecting my table from the page 
table[3].head()
```

```python

df=table[3]
```

```python
df.head(3)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'vte Club' : 'Club', 'Title(s)' : 'Titles_Won'}, inplace=True)

```

```python
df.head(2)
```

```python
# Calculate total titles
total_titles = df['Titles_Won'] + df['Runners-up']

# Calculate percentage of titles won and add it as a new column
df['Title_Won_Percentage'] = round((df['Titles_Won'] / total_titles) * 100,2)

# Calculate percentage of titles being the runner-up and add it as a new column
df['Runner_Up_Percentage'] =round((df['Runners-up'] / total_titles) * 100,2)
```

```python
df.head(2)
```

```python
# Calculate total titles
df['Total_Titles'] = df['Titles_Won'] + df['Runners-up']
grand_total_titles = df['Total_Titles'].sum()

# Calculate percentage of each category relative to grand total
df['Title_Won_Percentage_of_Grand_Total'] = round((df['Titles_Won'] / grand_total_titles) * 100, 2)
df['Runner_Up_Percentage_of_Grand_Total'] = round((df['Runners-up'] / grand_total_titles) * 100, 2)
```

```python
df.head(15)
```

```python
df.tail(15)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Club'], df['Titles_Won'], color='blue')
plt.xlabel('Club Name')
plt.ylabel('Number of Titles Won')
plt.title('Titles Won by Each Club')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
# Filter teams that have won the title at least once
df_won_once = df[df['Titles_Won'] >= 1]

# Plot teams that have won at least once
plt.figure(figsize=(8, 6))
plt.bar(df_won_once['Club'], df_won_once['Titles_Won'], color='blue')
plt.xlabel('Club')
plt.ylabel('Number of Titles Won')
plt.title('Teams That Have Won the Title at Least Once')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

The drama of the Champions League continues, especially after Real Madrid's incredible comeback. Inspired, I decided to check into past finals using Python's pandas library. Here's what the data reveals:

Real Madrid: Undisputed kings? They've reached a staggering 17 finals, winning 14! Talk about dominance and football heritage (pun was intended).


Real Madrid has taken this competition as their birth right must hurt to be a rival. Barcelona fans can bear witness to this. Milan which could have rivaled this are so off in terms of competiting at top level. Liverpool and Bayern Munich are pretty good in this competition

```python
# Filter teams that have been runner-up at least once
df_runner_up_once = df[df['Runners-up'] >= 1]

# Plot teams that have been runner-up at least once
plt.figure(figsize=(8, 6))
plt.bar(df_runner_up_once['Club'], df_runner_up_once['Runners-up'], color='orange')
plt.xlabel('Club')
plt.ylabel('Number of Times Runner-Up')
plt.title('Teams That Have Been Runner-Up at Least Once')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```



Atletico Madrid: Interesting stat! They've made it to 3 finals, but haven't clinched the trophy yet. they go again next year... And there are the likes of ARSENAL, Monaco, Roma and their likes who got there once and decided the competition wasn't made for them

Juventus has lost the most finals must really hurt to be a fan, Bayern Munich and Benfica are joint second with 5 finals lost but seems okay because they have won it a couple of times.

```python
# Select only the 'Titles_Won' and 'Runners-up' columns
df_selected = df[['Club', 'Titles_Won', 'Runners-up']]

# Set 'Club' as the index for plotting
df_selected.set_index('Club', inplace=True)

# Plot stacked bar chart
df_selected.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Club')
plt.ylabel('Count')
plt.title('Titles Won vs Runners-Up by Club')
plt.xticks(rotation=90)
plt.legend(title='Achievement')
plt.tight_layout()
plt.show()
```

Juventus: Champions League bridesmaids? They've been to 9 finals, but only secured victory twice. Ouch!
